```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/cleaned_crime_data_full.csv")
```

```python
print("Dataset Shape:", df.shape)
print(df.dtypes)
print(df.describe())
print(df.head())
print(df.tail())
```

```python
plt.figure(figsize=(8,5))
df.groupby('Hour').size().plot(kind='bar')
plt.title("Crime Frequency by Hour")
plt.xlabel("Hour")
plt.ylabel("Number of Crimes")
plt.show()
```

```python
df['Time_of_Day'].value_counts().plot(kind='bar')
plt.title("Crime by Time of Day")
plt.show()
```

```python
df['Area Name'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Crime Areas")
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.scatter(df['Longitude'], df['Latitude'], s=1)
plt.title("Crime Location Distribution")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

```python
df['Crime Description'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Crime Types")
plt.show()
sns.heatmap(df[['Hour','Month','Year','Latitude','Longitude']].corr(), annot=True)
plt.title("Correlation Matrix")
plt.show()
```
